### Indexing and Retrieving

A DataFrame can be indexed in a few different ways.

To get a single column, you can use a DataFrame['Name'] construction. Let's use this to answer a question about that column alone: what is the mean overall rating of all the players in our dataframe?

```python
data['Overall'].mean()
```

<b>Boolean indexing</b> with one column is also very convenient. The syntax is df[P(df['Name'])], where P is some logical condition that is checked for each element of the Name column. The result of such indexing is the DataFrame consisting only of rows that satisfy the P condition on the Name column.

Let’s use it to answer the question:<br>
What is the average rating for Club Manchester United?

```python
data[data.Club=='Manchester United']['Overall'].mean()
```

What is the average rating for players from Spain in Manchester United?

```python
data[(data.Club=='Manchester United') & (data.Nationality=='Spain')]['Overall'].mean()
```

DataFrames can be indexed by column name (label) or row name (index) or by the serial number of a row. The loc method is used for indexing by name, while iloc() is used for indexing by number.

In the first case below, we say "give us the values of the rows with index from 0 to 5 (inclusive) and columns labeled from Name to Potential (inclusive)". In the second case, we say "give us the values of the first five rows in the first three columns" (as in a typical Python slice: the maximal value is not included).

```python
data.loc[0:5, 'Name':'Potential']
```

```python
data.iloc[0:5,0:3]
```

If we need the first or the last line of the data frame, we can use the df[:1] or df[-1:] construct:

```python
data[:1]
```

### Grouping<br>
In general, grouping data in Pandas works as follows:

df.groupby(by=grouping_columns)[columns_to_show].function()<br>
<li>First, the groupby method divides the grouping_columns by their values. They become a new index in the resulting dataframe.</li>
<li>Then, columns of interest are selected (columns_to_show). If columns_to_show is not included, all non groupby clauses will be included.</li>
<li>Finally, one or several functions are applied to the obtained groups per selected columns.</li> <br>

Here's an example of getting club wise avg overall rating

```python
data.groupby(['Club'])['Overall'].mean().sort_values(ascending=False).head(10)
```

Let’s do the same thing, but slightly differently by passing a list of functions to agg():

```python
data.groupby(['Club'])['Overall'].agg([np.mean,np.max,np.min]).head(10)
```

### Summary tables
Suppose we want to see how the observations in our sample are distributed in the context of two variables - Churn and International plan. To do so, we can build a contingency table using the crosstab method:

```python
pd.crosstab(data['Nationality'], data['Club'])
```

Pivot tables are implemented in Pandas: the pivot_table method takes the following parameters:

values – a list of variables to calculate statistics for,<br>
index – a list of variables to group data by,<br>
aggfunc – what statistics we need to calculate for groups, ex. sum, mean, maximum, minimum or something else.<br>

Let's look at average of sprint speed, shooting, strength

```python
data.pivot_table(['SprintSpeed', 'ShotPower', 'Strength'],
               ['Club'], aggfunc='mean').head(10)
```

### DataFrame transformations
Like many other things in Pandas, adding columns to a DataFrame is doable in many ways.

For example, if we want to add a column which gives the difference in the overall and potential of a player, we can create it directly 

```python
data['gap_to_potential']=data['Potential']-data['Overall']
data.sort_values(by='gap_to_potential',ascending=False).head(10)
```

To delete columns or rows, use the drop method, passing the required indexes and the axis parameter (1 if you delete columns, and nothing or 0 if you delete rows). The inplace argument tells whether to change the original DataFrame. With inplace=False, the drop method doesn't change the existing DataFrame and returns a new one with dropped rows or columns. With inplace=True, it alters the DataFrame.

```python
data.drop(['Photo', 'Flag'], axis=1, inplace=True)
data.head(10)
```
